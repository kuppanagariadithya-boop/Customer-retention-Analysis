LOAD DATASET

```python
import pandas as pd

df = pd.read_csv("coustmer _data.CSV")
df.head()
```

DATA CLEANING

```python
df['LastPurchaseDate'] = pd.to_datetime(df['LastPurchaseDate'])

# Reference date (today)
today = pd.to_datetime("2024-01-01")

# Create Recency
df['Recency'] = (today - df['LastPurchaseDate']).dt.days
```

RFM ANALYSIS

```python
# Rename for clarity
df['Frequency'] = df['TotalPurchases']
df['Monetary'] = df['TotalSpent']

rfm = df[['CustomerID','Recency','Frequency','Monetary']]
print(rfm)
```

COUSTMER SEGMENTATION

```python
def segment(row):
    if row['Recency'] < 30 and row['Frequency'] > 5:
        return "Loyal"
    elif row['Recency'] < 60:
        return "Active"
    else:
        return "At Risk"

df['Segment'] = df.apply(segment, axis=1)
print(df[['CustomerID','Segment']])
```

CHURN ANALYSIS

```python
churn_count = df['Churn'].value_counts()
print(churn_count)
```

VIUALIZATION

```python
import matplotlib.pyplot as plt

# Churn Distribution
df['Churn'].value_counts().plot(kind='bar')
plt.title("Churn vs Retained Customers")
plt.show()

# Segment Distribution
df['Segment'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Customer Segments")
plt.show()
```
